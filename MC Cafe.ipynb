Your instructors love coffee so much that they decided to open a cafe! Of course, we will need the help of our students to 
develop the ordering app that the customers can use. On the cafe menu, there are three categories for the items we sell - 
coffee, tea, and pastries. When your program first starts, you must greet the customers by saying, "Welcome to MC Cafe!" Then 
ask the customer if they would like to see the menu. If yes, then display the menu items and prices. Then ask the user if they 
would like to order. If yes, then prompt the customer to enter an item and the quantity that they want. After they input an 
item, then ask them if they would like to order another item. Once your customers are finished with the order input, then 
display an order receipt that lists the items that they ordered and the quantity. Then at the bottom of the receipt, display 
the subtotal (the amount of the item before tax), the amount in sales tax (our business is in Maryland, so state tax is 6%), 
and finally their grand total (subtotal plus sales tax). At the end of their transaction, display a message to our customers 
saying, "Thank you for visiting MC Cafe!"

```python
print("Welcome to MC Cafe!")
subtotal = 0.0
shop_cart = {}
items = ['Espresso','Americano','Caffe latte','Mocha','Cappucino','Macchiato','White tea','Green tea','Oolong tea',
'Black tea','Earl Grey','Matcha','Croissant','Cinnamon roll','Banana bread','Pumpkin bread','Sticky bun','Apple turnover']
prices = [1.5,2.0,2.25,2.25,2.50,2.50,1.50,2.0,2.25,2.25,2.50,2.50,5.50,4.50,4.25,3.75,4.25,3.75]
menu = """
Coffee
1. Espresso - $1.50
2. Americano - $2.00
3. Caffe latte - $2.25
4. Mocha - $2.25
5. Cappucino - $2.50
6. Macchiato - $2.50
Tea
7. White tea - $1.50
8. Green tea - $2.00
9. Oolong tea - $2.25
10 Black tea - $2.25
11 Earl Grey - $2.50
12 Matcha - $2.50
Pastries
13 Croissant - $5.50
14 Cinnamon roll - $4.50
15 Banana bread - $4.25
16 Pumpkin bread - $3.75
17 Sticky bun - $4.25
18 Apple turnover - $3.75"""
resp = input("Would you like to see the menu?[y|n] ")
if resp=='y':
    print(menu)
    resp = input("Would you like to order?[y|n] ")
    if resp=='y':
        while resp=='y':
            itemNo = int(input("Enter the desired item number: "))
            quant = int(input("Enter the quantity desired: "))
            shop_cart[items[itemNo-1]] = quant
            subtotal += prices[itemNo-1] * quant   # Menu index starts with 1 not 0 as in prices[]
            tax = .06 * subtotal
            resp = input("Would you like to order another item?[y|n] ")
        print("\nYour Order:")    
        for item, count in shop_cart.items():
            print(item,".....",count)
        print("Subtotal = ","${:.2f}".format(subtotal))     
        print("Tax = ","${:.2f}".format(tax))
        print("Grand Total = ", "${:.2f}".format(subtotal + tax))
        print("Thank you for visiting MC Cafe!") 
    else:
        print("Thank you for visiting MC Cafe!")
else:
    print("Thank you for visiting MC Cafe!")
```
